```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
  
CTS = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
```

```python
print(CTS)
CTS
```

```python
#PRIVATE & PUBLIC CONSTRUCTION SPENDING

#Set x and y axis(s)
x1 = CTS['Month-Year']
y1 = CTS['Total Construction']

#Creating figure and lines
fig,ax = plt.subplots()
ax.plot(x1,y1,color='b',linewidth=1)


#Setting parameters of graph
ax.grid(False)
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Month-Year',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Spending (in $ amount)',fontsize='14',fontname='Times New Roman')
ax.set_xticks(x1[::12])
ax.set_xticklabels(x1[::12])
ax.set_ylim(0,max(y1)+2000)

""" Enhance figure object """
fig.set_size_inches(15,8)
fig.suptitle("Total Construction Spending 2002-2014",fontsize='18',fontname='Times New Roman')
```
